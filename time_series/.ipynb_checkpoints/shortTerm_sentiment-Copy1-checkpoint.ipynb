```python
def featureExtraction(usersFrequency, label):
    X = []
    emotions = ['surprise', 'fear', 'sadness', 'disgust', 'trust', 'anticipation', 'anger','joy']
    for userData in usersFrequency.values():
        userFeature = []
        for emotion in emotions:
            userFeature  += userData.get(emotion, emptyHourTable())
        X.append(userFeature)
    Y = np.zeros(len(usersFrequency), dtype=int)
    Y += label
    return np.array(X), Y
    
```

```python
colorTable = {"joy":"#FADB4D","sadness": "#729DC9","fear":"#35A450", "anticipation": "#F2993A", "anger":"#E43054", "trust": "#99CC33", "disgust": "#9F78BA" ,"surprise" : "#3FA5C0"}.keys()
```

```python
x1, y1 = featureExtraction(regularUserFrequency, 0)
x2, y2 = featureExtraction(BPDUserFrequency, 1)
X = np.concatenate((x1,x2))
Y = np.concatenate((y1,y2))
```

```python
forest = RandomForestClassifier(n_jobs= -1)
forest.fit(X,Y)
```

```python
def trainTestGenerator(regularUserFrequency, BPDUserFrequency, randomState):
    x1, y1 = featureExtraction(regularUserFrequency, 0)
    x2, y2 = featureExtraction(BPDUserFrequency, 1)
    x1_train,x1_test, y1_train, y1_test = train_test_split(x1,y1, random_state=randomState)
    x2_train,x2_test, y2_train, y2_test = train_test_split(x2,y2, random_state=randomState)
    
    X_train = np.concatenate((x1_train, x2_train))
    X_test = np.concatenate((x1_test, x2_test))
    Y_train = np.concatenate((y1_train, y2_train))
    Y_test = np.concatenate((y1_test, y2_test))
    
    return X_train, X_test, Y_train, Y_test
```

```python
for i in range(10):
    X_train, X_test, Y_train, Y_test = trainTestGenerator(regularUserFrequency, BPDUserFrequency, randomState = i)
    forest = RandomForestClassifier(n_jobs=-1, max_features="sqrt")
    forest.fit(X_train, Y_train)
    print(forest.score(X_test, Y_test))
```

```python
def plot_ROC(classifier, X_train, X_test, Y_train, Y_test ):
    classifier.fit(X_train, Y_train)
    scores = classifier.predict_proba(X_test)
    

    fpr, tpr, _ = roc_curve(Y_test, scores[:,1])
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

```python
X_train, X_test, Y_train, Y_test = trainTestGenerator(regularUserFrequency, BPDUserFrequency, randomState = 8)
plot_ROC(RandomForestClassifier(n_jobs=-1, max_features="sqrt",n_estimators=128), X_train, X_test, Y_train, Y_test)
```

```python
X_train, X_test, Y_train, Y_test = trainTestGenerator(regularUserFrequency, BPDUserFrequency, randomState = 12)
plot_ROC(SGDClassifier(loss="log", shuffle=True), X_train, X_test, Y_train, Y_test)
```
